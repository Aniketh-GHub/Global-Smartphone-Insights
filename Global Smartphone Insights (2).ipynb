```python
#While working on EDA (Exploratory Data Analysis) projects, we typically follow certain steps. 
#The initial step involves collecting the data.
#Then the task goes on
#Step-1 : Importing libraries 
import mysql.connector
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import math
```

```python
#Step - 2: Connecting the SQL database 
Pdata = mysql.connector.connect(
    host = "localhost",
    user= "root",
    password = "123456789",
    database = "phones",
    auth_plugin="mysql_native_password"
)
mycursor=Pdata.cursor()
print(Pdata)
```

```python
#Step-2: Loading the data
mycursor.execute("Select * from mobiles")
output = mycursor.fetchall()
for x in output :
    print(x)
```

```python
#Step - 3 : Retrieve data
query = "Select * from mobiles"
Pdata = pd.read_sql(query,Pdata)
```

```python
#Here we have any null values, so we must replace them with their corresponding mean values
```

```python
#Calculating the mean value for cost
Mean_Cost= int(Pdata['Cost_INR'].mean())
Mean_Segment = math.ceil(Pdata['segment'].mean())
print("Mean data of the prices: ",Mean_Cost)
print("Mean data of the segment: ",Mean_Segment)
```

```python
# Replace NaN values in 'Cost_INR' and 'segment' with the calculated mean
# Fill NaN values in 'Cost_INR' and 'segment' columns with the mean, modifying the original DataFrame
Pdata['Cost_INR'].fillna(Mean_Cost, inplace=True)
Pdata
print(Pdata)
Pdata['segment'].fillna(Mean_Segment, inplace=True)
Pdata
print(Pdata)
```

```python
#We check again for null values 
Pdata.isna().sum()
```

```python
#We cleaned the columns containing the NULL values
#Now we must check for the duplicate values
duplicates = Pdata[Pdata.duplicated()]
# Display columns containing duplicate rows
print(duplicates)
```

```python
#Since we got duplicates, we must remove them 
dup=Pdata.drop_duplicates(inplace=True)
print(dup)
```

```python
#We removed all the duplicate values
```

```python
# Step-5 : We will now analyze the data
#Getting histogram based on the values 
Pdata.hist(figsize=(12, 10), bins=20)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(Pdata['Charger_volt'], bins=30, kde=True)
plt.title('Distribution of Charger_volt')
plt.xlabel('Chargers (in volts)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Boxplot to detect outliers in numerical features
plt.figure(figsize=(12, 8))
sns.boxplot(data=Pdata[['ram', 'rom', 'Battery', 'Cost_INR']])
plt.show()
```

```python
plt.figure(figsize=(12, 8))
Pdata['segment'] = pd.Categorical(Pdata['segment'], categories=[0, 1, 2, 3], ordered=True)
sns.barplot(x='segment', y='Cost_INR', data=Pdata)
plt.title('Phone prices corresponding to their segments ')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='ram', y='rom', data=Pdata)
plt.title('RAM vs ROM')
plt.xlabel('RAM')
plt.ylabel('ROM')
plt.show()
```

```python
sns.pairplot(Pdata[['front_camera', 'back_camera', 'Battery', 'segment']])
plt.suptitle('Pair Plot of Metrics', y=1.02)
plt.show()
```

```python
# Plot pie chart for Processors
# Count occurrences of Processors
display_counts = Pdata['display'].value_counts().head()
plt.figure(figsize=(8, 6))
plt.pie(display_counts, labels= display_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of display type(in inches)')
plt.show()
```

```python
# Exclude non-numeric columns for correlation matrix and pair plot
numeric_data = Pdata[['ram', 'rom', 'display', 'front_camera', 'back_camera', 'Battery', 'Charger_volt', 'Cost_INR']]
plt.figure(figsize=(10, 8))
corr_matrix = numeric_data.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
#Scatterplotting for Cost and Ram data
plt.figure(figsize=(12, 6))
sns.scatterplot(x='ram', y='Cost_INR', data=Pdata, color='green')
plt.title('Cost vs RAM')
plt.xlabel('RAM (GB)')
plt.ylabel('Cost (INR)')
plt.show()
```

```python
#Scatterplotting for Cost and Rom data
plt.figure(figsize=(12, 6))
sns.scatterplot(x='rom', y='Cost_INR', data=Pdata, color='red')
plt.title('Cost vs ROM')
plt.xlabel('ROM (GB)')
plt.ylabel('Cost (INR)')
plt.show()
```
